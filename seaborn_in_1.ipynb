```python
# Advanced visualization using seaborn
import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

df
```

```python
count = df['# Gears'].value_counts()
print(count)
#print(count.plot(kind='pie'))
count.plot(kind='bar')
```

```python
import seaborn as sns
sns.set() # to set parameters that controls general plots
count.plot(kind='pie')
```

```python
import seaborn as sns
sns.set()
count.plot(kind='bar')
```

```python
# relationship between cylinders, city MPG, high MPG, and combined MPG

df2 = df[['Cylinders','CityMPG','HwyMPG','CombMPG']]
print(type(df2))
print(df2.shape)
```

```python
df2.head(10)
```

```python
sns.pairplot(df2,height=3)
```

```python
# combined milleage scatter plot

sns.scatterplot(x='Cylinders',y='CombMPG',data=df2)
```

```python
sns.jointplot(x='Cylinders',y='CityMPG',data=df2)
```

```python
#LM plot
sns.lmplot(x='Cylinders',y='CombMPG',data=df2) # lmplot = linear model plot / line manipulation facet grid plot
```

```python
sns.jointplot(x='Cylinders',y='HwyMPG',data=df2)
```

```python
# whisker plot
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.boxplot(x='Mfr Name',y='CombMPG',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
# swarmp plot
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.swarmplot(x='Mfr Name',y='CombMPG',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
# count plot
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(x='Mfr Name',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```

```python
# heatmap
df2 = df.pivot_table(index='Cylinders',columns = 'Eng Displ',values = 'CityMPG',aggfunc='mean')
sns.heatmap(df2) # data mapping is represented in color
```
